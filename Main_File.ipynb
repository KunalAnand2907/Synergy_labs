## 1.) Creating a Directory and Saving the extracted Frames from the Video

```python
# IMPORTING ESSENTIAL LIBRARIES
import os
import numpy as np
import cv2
from glob import glob

# Function 1: to create a Directory by name save_frames, will have all the extracted frames in it
def create_dir(path):
    try:
        if not os.path.exists(path):
            os.makedirs(path)
    except OSError:
        print(f" ERROR: creating  directory with name (path)")

# Function 2: to create a function which extracts frames and save it by Frame.Id.jpg 
def save_frame(video_path, save_dir):
    name = video_path[0].split("\\")[-1].split(".")[0]# Getting the video name 
    save_path = os.path.join(save_dir, name) 
    create_dir(save_path) # calling the function
    
    cap = cv2.VideoCapture(video_path)
    id = 0
    
    while True:
        ret, frame = cap.read()
        if ret == False:
            cap.release()
            break
        else:
            cv2.imwrite(r"C:\Users\KUNAL\Desktop\Synergy\save_frames\Frame." +str(id)+ ".jpg",frame)
            cv2.imshow("frame",frame)
            id +=1
            
        

if __name__ == "__main__": # if source file run directly from main, not imported from another file dir
    
    video_paths = glob("videos/*") # Gives a list of all files match w.r.t a pattern given 
    save_dir = "save_frames"
    
    for path in video_paths:
        save_frame(path, save_dir)

```

## 2.) Creating a directory and extracting the Frames and Rotating by 180 Degree from the Video

```python
# IMPORTING ESSENTIAL LIBRARIES
import os
import numpy as np
import cv2
from glob import glob


def create_dir(path):
    try:
        if not os.path.exists(path):
            os.makedirs(path)
    except OSError:
        print(f" ERROR: creating  directory with name (path)")

# Function 1: 
def save_frame(video_path, save_dir):
    name = video_path[0].split("\\")[-1].split(".")[0]
    save_path = os.path.join(save_dir, name)
    create_dir(save_path)
    
    cap = cv2.VideoCapture(video_path)
    id = 0
    
    while True:
        ret, frame = cap.read()
        if ret == False:
            cap.release()
            break
        else:
            images=cv2.rotate(frame,cv2.ROTATE_180)
            cv2.imwrite(r"C:\Users\KUNAL\Desktop\Synergy\rotated_frames\Frame."+str(id)+".jpg",images)
            id +=1
        

if __name__ == "__main__": # if source file run directly from main, not imported from another file dir
    
    video_paths = glob("videos/*") # Gives a list of all files match w.r.t a pattern given 
    save_dir = "rotated_frames"
    
    for path in video_paths:
        save_frame(path, save_dir)
```
